```python
def year_division(year):
    if year == 0:
        return "Date unknown"
    elif year <= 1800:
        return "<=1800"
    elif year <= 1850:
        return "1801-1850"
    elif year <= 1900:
        return "1851-1900"
    elif year <= 1950:
        return "1901-1950"
    elif year <= 2000:
        return "1951-2000"
    else:
        return "2000>"

df["YEAR BUILT GROUP"] = df["YEAR BUILT"].apply(lambda x: year_division(x))
df
```

```python
df.groupby("YEAR BUILT GROUP").count().plot.bar(y="BOROUGH", xlabel="Time when wasbuilt", ylabel="Number of buildings")
```

```python
df.groupby("YEAR BUILT GROUP").mean().plot.bar(y="SALE\nPRICE", xlabel="Time when wasbuilt", ylabel="Sale price")
```
